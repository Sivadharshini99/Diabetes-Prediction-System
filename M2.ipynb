# Data visualization:

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
```

```python
df = data.dropna()
```

```python
df.columns
```

```python
sns.countplot(x="diabetes", data=df, palette="husl")
```

```python
df['cholesterol'].hist(figsize=(5,5), color='blue', alpha=0.7)
plt.xlabel('cholesterol')
plt.ylabel('Count')
plt.title('cholesterol and its Count')
```

```python
df['systolic_bp'].hist(figsize=(5,5), color='yellow', alpha=1)
plt.xlabel('systolic_bp')
plt.ylabel('Count')
plt.title('systolic_bp and its Count')
```

```python
def piechart(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)

piechart(df, 'diabetes')
```

```python
def piechart(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)

piechart(df, 'gender')
```

```python
g = sns.catplot("diabetes", data=df, kind='count', hue='gender', margin_titles=False)
g.set_ylabels('Number of datas')
g.fig.set_figheight(5)
g.fig.set_figwidth(20)
```

```python
sns.scatterplot(df.glucose, df.cholesterol)
```

```python
sns.pairplot(df)
```

```python
fig, ax = plt.subplots(figsize=(25,12))
sns.heatmap(df.corr(),annot = True, ax=ax)
plt.show()
```
